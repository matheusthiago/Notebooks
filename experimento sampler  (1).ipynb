```r
min=min(dado)
max=max(dado)
norm.dado=(dado-min)/(max-min)
min(norm.dado)
#troca do zero por 0.00001, para não dar problema com o MAPE, que não aceita valores negativos e zero
consumption=replace(x = norm.dado, list = norm.dado==0, values = 0.00001)
plot(consumption)
```
