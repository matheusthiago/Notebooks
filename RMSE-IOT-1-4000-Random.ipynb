```r
#visualização de RMSE no jupyter
library(ggplot2)
mal=as.data.frame(read.table("IOT/random/grouped_mal.txt"))
kal=as.data.frame(read.table("IOT/random/grouped_kal.txt"))
sti=as.data.frame(read.table("IOT/random/grouped_sti.txt"))
spl=as.data.frame(read.table("IOT/random/grouped_spl.txt"))
ssa=as.data.frame(read.table("IOT/random/grouped_ssa.txt"))

names(mal)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(kal)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(sti)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(spl)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(ssa)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")


#Tirando a média 
tamanhoGap= seq(1,4000,10) 
mediaSpl=c()
mediaKal=c()
mediaMal=c()
mediaSti=c()
mediaSsa=c()

sdSpl=c()
sdKal=c()
sdMal=c()
sdSti=c()
sdSsa=c()

sdLacSpl=c()
sdLacKal=c()
sdLacMal=c()
sdLacSti=c()
sdLacSsa=c()


mediaLacSpl=c()
mediaLacKal=c()
mediaLacMal=c()
mediaLacSti=c()
mediaLacSsa=c()

a=c()

for (i in seq(1,400,1)) {
    #calculando a média
	mediaSsa=append(mediaSsa,(mean(ssa$RMSE[seq(i,12000, 400)])))
	mediaSpl=append(mediaSpl,(mean(spl$RMSE[seq(i,12000, 400)])))
	mediaKal=append(mediaKal,(mean(kal$RMSE[seq(i,12000, 400)])))
	mediaMal=append(mediaMal,(mean(mal$RMSE[seq(i,12000, 400)])))
	mediaSti=append(mediaSti,(mean(sti$RMSE[seq(i,12000, 400)])))

    #Calculando o desvio padrão
    sdSsa=append(sdSsa,(sd(ssa$RMSE[seq(i,12000, 400)])))
	sdSpl=append(sdSpl,(sd(spl$RMSE[seq(i,12000, 400)])))
	sdKal=append(sdKal,(sd(kal$RMSE[seq(i,12000, 400)])))
	sdMal=append(sdMal,(sd(mal$RMSE[seq(i,12000, 400)])))
	sdSti=append(sdSti,(sd(sti$RMSE[seq(i,12000, 400)])))
}

mediaKal=as.data.frame(mediaKal)
names(mediaKal)=c("media")
mediaKal$sd=sdKal
mediaKal$tamanho=tamanhoGap
mediaKal$algoritmo="Kalman"

mediaMal=as.data.frame(mediaMal)
names(mediaMal)=c("media")
mediaMal$sd=sdMal
mediaMal$tamanho=tamanhoGap
mediaMal$algoritmo="Moving Average"


mediaSti=as.data.frame(mediaSti)
names(mediaSti)=c("media")
mediaSti$sd=sdSti
mediaSti$tamanho=tamanhoGap
mediaSti$algoritmo="Stine"

mediaSpl=as.data.frame(mediaSpl)
names(mediaSpl)=c("media")
mediaSpl$sd=sdSpl
mediaSpl$tamanho=tamanhoGap
mediaSpl$algoritmo="Spline"

mediaSsa=as.data.frame(mediaSsa)
names(mediaSsa)=c("media")
mediaSsa$sd=sdSsa
mediaSsa$tamanho=tamanhoGap
mediaSsa$algoritmo="Ssa"

todos=rbind(mediaKal,mediaMal, mediaSsa, mediaSti, mediaSpl)
```
