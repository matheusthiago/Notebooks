```r
#mu= expeted return of the stock per annum
#sigma annualizes volatility
rwalk = function(mu, sigma){
    #inicial price
    p0=1000
    days = 250
    mu.daily=mu/days
    sigma.daily = sigma *1/sqrt(days)
    
    r=rnorm(100,mu.daily,sigma.daily) #100 random normals
    logPrice=log(p0) + cumsum(r) #reutns a vector whose elements are the cumulative sums of the arguments elements
    prices= exp(logPrice)
    
}
```
