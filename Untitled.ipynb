```r
x =x0= dadoDia
piso=7000
teto=8000
x[piso:teto] <- NA
fit <- auto.arima(x)
#fit1 <- forecast(Arima(AirPassengers[1:4100],model=fit),h=10)
#fit2 <- forecast(Arima(rev(AirPassengers[4600:length(x)]), model=fit), h=10)
kr <- KalmanRun(x, fit$model)
y <- x
# impute missing values Z %*% alpha at each missing observation
id.na <- which(is.na(x))

for (i in id.na)
  y[i] <- fit$model$Z %*% kr$states[i,]
# alternative to the explicit loop above
sapply(id.na, FUN = function(x, Z, alpha) Z %*% alpha[x,], 
  Z = fit$model$Z, alpha = kr$states)
#y[id.na]

par(mfrow = c(2, 1), mar = c(2.2,2.2,2,2))
plot(x0, col = "gray")
lines(x)
points(time(x0)[id.na], x0[id.na], col = "blue", pch = 19)
points(time(y)[id.na], y[id.na], col = "red", pch = 17)
legend("topleft", legend = c("true values", "imputed values"), 
  col = c("blue", "red"), pch = c(19, 17))
plot(time(x0)[piso:teto], x0[piso:teto], type = "b", col = "blue", 
  pch = 19, ylim = range(x0[60:71]))
points(time(y)[piso:teto], y[piso:teto], col = "red", pch = 17)
lines(time(y)[piso:teto], y[piso:teto], col = "red")
legend("topleft", legend = c("true values", "imputed values"), 
  col = c("blue", "red"), pch = c(19, 17), lty = c(1, 1))

rmse(xo,x)

#lines(ts(0.5*c(fit1$mean+rev(fit2$mean)), 
#  start=time(AirPassengers)[90],freq=12), col="red")
```

```r
options(rep.plot.width=10, rep.plot.heigth=40)
dfDados=readRDS("plumber/dfDadosIoTa.norm.rds")
dia=1
piso=4000
teto=5100
	dadoDia=dfDados$dados[which(dfDados$dia==dia)]
	dadoDiaTemp=dadoDia
	dadoDiaTemp[piso:teto]=NA #cria espaços em branco
	par(mfrow = c(5,1), oma = c(1,1,1,1))
    id.na <- which(is.na(dadoDiaTemp))
n=teto-piso

    plot(dadoDia,  main="Original")

    dadoImputado=na_interpolation(dadoDiaTemp)
	plot(dadoDiaTemp,  main="interpolation")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])




    dadoImputado=na_kalman(dadoDiaTemp, model="auto.arima")
	plot(dadoDiaTemp,  main="kalman Arima")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)


rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_locf(dadoDiaTemp)
	plot(dadoDiaTemp,  main="Locf")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])



    dadoImputado=na_ma(dadoDiaTemp, weighting="simple", k=n)
	plot(dadoDiaTemp,  main="SMA")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])

    dadoImputado=na_ma(dadoDiaTemp, weighting="linear", k=n)
	plot(dadoDiaTemp,  main="LMA")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_ma(dadoDiaTemp, weighting="exponential", k=n)
	plot(dadoDiaTemp,  main="EMA")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_mean(dadoDiaTemp)
	plot(dadoDiaTemp,  main="media")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])

    dadoImputado=na_mean(dadoDiaTemp, option="median")
	plot(dadoDiaTemp,  main="median")
   points(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])
```
