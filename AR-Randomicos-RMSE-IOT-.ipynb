Leitura e análise dos resultados dos experimentos das lacunas seguidas:
-----------
* 1 a 100 (com granularidade 1) 
* 150 a 2100 (com granularidade 50)

```r
#visualização de RMSE no jupyter
library(ggplot2)
library(tidyverse)
library(grid)
```

```r
#visualização de RMSE no jupyter
library(ggplot2)
mal1_4000=as.data.frame(read.table("IOT/random/grouped_mal.txt"))
kal1_4000=as.data.frame(read.table("IOT/random/grouped_kal.txt"))
sti1_4000=as.data.frame(read.table("IOT/random/grouped_sti.txt"))
spl1_4000=as.data.frame(read.table("IOT/random/grouped_spl.txt"))
ssa1_4000=as.data.frame(read.table("IOT/random/grouped_ssa.txt"))

names(mal1_4000)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(kal1_4000)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(sti1_4000)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(spl1_4000)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(ssa1_4000)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")


#Tirando a média 
tamanhoGap= seq(101,4000,10) 
mediaSpl1_4000=c()
mediaKal1_4000=c()
mediaMal1_4000=c()
mediaSti1_4000=c()
mediaSsa1_4000=c()

sdSpl1_4000=c()
sdKal1_4000=c()
sdMal1_4000=c()
sdSti1_4000=c()
sdSsa1_4000=c()

sdLacSpl1_4000=c()
sdLacKal1_4000=c()
sdLacMal1_4000=c()
sdLacSti1_4000=c()
sdLacSsa1_4000=c()


mediaLacSpl1_4000=c()
mediaLacKal1_4000=c()
mediaLacMal1_4000=c()
mediaLacSti1_4000=c()
mediaLacSsa1_4000=c()

a=c()

for (i in seq(11,400,1)) {
    #calculando a média
	mediaSsa1_4000=append(mediaSsa1_4000,(mean(ssa1_4000$RMSE[seq(i,12000, 400)])))
	mediaSpl1_4000=append(mediaSpl1_4000,(mean(spl1_4000$RMSE[seq(i,12000, 400)])))
	mediaKal1_4000=append(mediaKal1_4000,(mean(kal1_4000$RMSE[seq(i,12000, 400)])))
	mediaMal1_4000=append(mediaMal1_4000,(mean(mal1_4000$RMSE[seq(i,12000, 400)])))
	mediaSti1_4000=append(mediaSti1_4000,(mean(sti1_4000$RMSE[seq(i,12000, 400)])))

    #Calculando o desvio padrão
    sdSsa1_4000=append(sdSsa1_4000,(sd(ssa1_4000$RMSE[seq(i,12000, 400)])))
	sdSpl1_4000=append(sdSpl1_4000,(sd(spl1_4000$RMSE[seq(i,12000, 400)])))
	sdKal1_4000=append(sdKal1_4000,(sd(kal1_4000$RMSE[seq(i,12000, 400)])))
	sdMal1_4000=append(sdMal1_4000,(sd(mal1_4000$RMSE[seq(i,12000, 400)])))
	sdSti1_4000=append(sdSti1_4000,(sd(sti1_4000$RMSE[seq(i,12000, 400)])))
}

mediaKal1_4000=as.data.frame(mediaKal1_4000)
names(mediaKal1_4000)=c("media")
mediaKal1_4000$sd=sdKal1_4000
mediaKal1_4000$tamanho=tamanhoGap
mediaKal1_4000$algoritmo="Kalman"

mediaMal1_4000=as.data.frame(mediaMal1_4000)
names(mediaMal1_4000)=c("media")
mediaMal1_4000$sd=sdMal1_4000
mediaMal1_4000$tamanho=tamanhoGap
mediaMal1_4000$algoritmo="Moving Average"


mediaSti1_4000=as.data.frame(mediaSti1_4000)
names(mediaSti1_4000)=c("media")
mediaSti1_4000$sd=sdSti1_4000
mediaSti1_4000$tamanho=tamanhoGap
mediaSti1_4000$algoritmo="Stine"

mediaSpl1_4000=as.data.frame(mediaSpl1_4000)
names(mediaSpl1_4000)=c("media")
mediaSpl1_4000$sd=sdSpl1_4000
mediaSpl1_4000$tamanho=tamanhoGap
mediaSpl1_4000$algoritmo="Spline"

mediaSsa1_4000=as.data.frame(mediaSsa1_4000)
names(mediaSsa1_4000)=c("media")
mediaSsa1_4000$sd=sdSsa1_4000
mediaSsa1_4000$tamanho=tamanhoGap
mediaSsa1_4000$algoritmo="Ssa"

```

```r
#lendo os arquivos dos experimentos dos experimentos de 1 a 100 de lacunas seguidas 
#visualização de RMSE no jupyter
library(ggplot2)
mal1_100=as.data.frame(read.table("IOT/random1-100/grouped_mal.txt"))
kal1_100=as.data.frame(read.table("IOT/random1-100/grouped_kal.txt"))
sti1_100=as.data.frame(read.table("IOT/random1-100/grouped_sti.txt"))
spl1_100=as.data.frame(read.table("IOT/random1-100/grouped_spl.txt")) 


names(mal1_100)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(kal1_100)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(sti1_100)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")
names(spl1_100)=c("task-clock", "faults", "cs", "migrations","time", "RMSE")


#Tirando a média 
tamanhoGap= seq(1,100,1) 
mediaKal1_100=c()
mediaMal1_100=c()
mediaSti1_100=c()
mediaSpl1_100=c()

sdKal1_100=c()
sdMal1_100=c()
sdSti1_100=c()
sdSpl1_100=c()

sdLacKal1_100=c()
sdLacMal1_100=c()
sdLacSti1_100=c()
sdLacSpl1_100=c()

mediaLacKal1_100=c()
mediaLacMal1_100=c()
mediaLacSti1_100=c()
mediaLacSpl1_100=c()

a=c()

for (i in seq(1,100,1)) {
    #calculando a média
	mediaKal1_100=append(mediaKal1_100,(mean(kal1_100$RMSE[seq(i,3000, 100)])))
	mediaMal1_100=append(mediaMal1_100,(mean(mal1_100$RMSE[seq(i,3000, 100)])))
	mediaSti1_100=append(mediaSti1_100,(mean(sti1_100$RMSE[seq(i,3000, 100)])))
	mediaSpl1_100=append(mediaSpl1_100,(mean(spl1_100$RMSE[seq(i,3000, 100)])))

    #Calculando o desvio padrão
	sdKal1_100=append(sdKal1_100,(sd(kal1_100$RMSE[seq(i,3000, 100)])))
	sdMal1_100=append(sdMal1_100,(sd(mal1_100$RMSE[seq(i,3000, 100)])))
	sdSti1_100=append(sdSti1_100,(sd(sti1_100$RMSE[seq(i,3000, 100)])))
	sdSpl1_100=append(sdSpl1_100,(sd(spl1_100$RMSE[seq(i,3000, 100)])))
}

mediaKal1_100=as.data.frame(mediaKal1_100)
names(mediaKal1_100)=c("media")
mediaKal1_100$sd=sdKal1_100
mediaKal1_100$tamanho=tamanhoGap
mediaKal1_100$algoritmo="Kalman"

mediaMal1_100=as.data.frame(mediaMal1_100)
names(mediaMal1_100)=c("media")
mediaMal1_100$sd=sdMal1_100
mediaMal1_100$tamanho=tamanhoGap
mediaMal1_100$algoritmo="Moving Average"


mediaSti1_100=as.data.frame(mediaSti1_100)
names(mediaSti1_100)=c("media")
mediaSti1_100$sd=sdSti1_100
mediaSti1_100$tamanho=tamanhoGap
mediaSti1_100$algoritmo="Stine"

mediaSpl1_100=as.data.frame(mediaSpl1_100)
names(mediaSpl1_100)=c("media")
mediaSpl1_100$sd=sdSpl1_100
mediaSpl1_100$tamanho=tamanhoGap
mediaSpl1_100$algoritmo="Spline"
```

```r
todos=rbind(mediaKal1_100,
            mediaMal1_100,
            mediaSti1_100,
            mediaSpl1_100,
          #  mediaSsa1_100,
            mediaKal1_4000,
            mediaMal1_4000,
            mediaSti1_4000,
            mediaSpl1_4000
           )
summary(todos)
```
