```r
options(rep.plot.ts.width=10, rep.plot.ts.heigth=40)
dfDados=readRDS("../plumber/dfDadosIoTa.norm.rds")
dia=1
piso=4000
teto=5100
	dadoDia=dfDados$dados[which(dfDados$dia==dia)]
	dadoDiaTemp=dadoDia
	dadoDiaTemp[piso:teto]=NA #cria espaços em branco
	par(mfrow = c(2,1), oma = c(1,1,1,1))
    id.na <- which(is.na(dadoDiaTemp))
n=teto-piso

    plot.ts(dadoDia,  main="Original")

    dadoImputado=na_interpolation(dadoDiaTemp)
	plot.ts(dadoDiaTemp,  main="interpolation")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])




    dadoImputado=na_kalman(ts(dadoDiaTemp, frequency = 10), model="auto.arima")
    plotNA.imputations(ts(dadoDia), dadoImputado)


rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_locf(dadoDiaTemp)
	plot.ts(dadoDiaTemp,  main="Locf")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])



    dadoImputado=na_ma(dadoDiaTemp, weighting="simple", k=n)
	plot.ts(dadoDiaTemp,  main="SMA")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])

    dadoImputado=na_ma(dadoDiaTemp, weighting="linear", k=n)
	plot.ts(dadoDiaTemp,  main="LMA")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_ma(dadoDiaTemp, weighting="exponential", k=n)
	plot.ts(dadoDiaTemp,  main="EMA")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_mean(dadoDiaTemp)
	plot.ts(dadoDiaTemp,  main="media")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])

    dadoImputado=na_mean(dadoDiaTemp, option="median")
	plot.ts(dadoDiaTemp,  main="median")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])

    s=ssa(dadoDiaTemp, L=(teto-piso))
    dadoImputado=gapfill(s, groups=list(1:100))
	plot.ts(dadoDiaTemp,  main="ssa")
   lines(time(dadoImputado)[id.na], dadoImputado[id.na], col = "red", pch = 17)
rmse(dadoDia[piso:teto], dadoImputado[piso:teto])


    dadoImputado=na_seadec(ts(dadoDiaTemp, frequency = 360))
    plotNA.imputations(ts(dadoDia), dadoImputado)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])
 

    dadoImputado=na_seasplit(ts(dadoDiaTemp, frequency = 360))
    plotNA.imputations(ts(dadoDia), dadoImputado)

rmse(dadoDia[piso:teto], dadoImputado[piso:teto])
```
